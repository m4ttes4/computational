```julia
using DataFrames, CSV
```

```julia
df = CSV.File("densit.dat", comment="#", delim="  ",
    normalizenames=false, header=1) |> DataFrame
#df = DataFrame(CSV.File("pressure_winds.dat"))
select!(df, Not(" #"))
```

```julia
eachcol(df)
```

```julia
x = [df.Column2]
y = [df.Column3]
using Plots

plotlyjs()
plot(x,y,
    label = "credo sia la densità",
    title="titolo"
    )
plot!(xscale=:log10,
    yscale=:log10)
xlabel!("distance")
ylabel!("density")
```
